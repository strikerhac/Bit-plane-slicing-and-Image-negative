```python
import numpy
import matplotlib
from PIL import Image
from scipy import misc
import numpy as np
import matplotlib.pyplot as plt
```

```python
def rgb2gray2binary(url):
    im = Image.open(url).convert("L")
    im.save("gray.png")
    im = Image.open("gray.png")
    im = im.point(lambda x: 0 if x<128 else 255)
    im.save("binary.png")
```

```python
rgb2gray2binary("lena_color.gif")
```

```python
def image_matrix(url):
    image = misc.imread(url)
    return image
```

```python
def image_negative(url, image_mode):
    image = image_matrix(url)
    for r in range(len(image)): 
        for c in range(len(image[r])):
            image[r][c] = 255 - image[r][c]
    if image_mode == "rgb":
        plt.imshow(image)
    if image_mode == "g":
        plt.imshow(image, cmap = matplotlib.cm.Greys_r)
    if image_mode == "b":
        plt.imshow(image, cmap = matplotlib.cm.binary_r)
```

```python
image_negative("lena_color.gif","rgb")
```

```python
image_negative("gray.png","g")
```

```python
image_negative("binary.png","b")
```

```python
def image_vertical_edges(url):
    im = Image.open(url)
    grad = np.zeros(im.size)
    arr = np.array(im)
    
    for r in range(len(arr)): 
        for c in range(len(arr[r])):
            if c == 0:
                continue
            elif c == 511:
                continue
            else:
                grad[r][c] = int(arr[r][c+1]) - int(arr[r][c-1])
    mini = np.min(grad)
    maxi = np.max(grad)
    
    for r,c in np.ndindex(grad.shape):
        grad[r][c] = ((grad[r][c]-mini)/(maxi - mini)) * 255
    
    plt.imshow(grad, cmap = matplotlib.cm.Greys_r)
```

```python
image_vertical_edges("gray.png")
```

```python
def bit_slicing(url, bit_plane):
    b = image_matrix(url)
    b_bits = np.unpackbits(b).reshape((b.shape)[0],(b.shape)[1],8)
    b_bits[:,:,:(8-bit_plane)] = 0
    b_aug = np.packbits(b_bits).reshape(b.shape)
    plt.imshow(b_aug, cmap = matplotlib.cm.Greys_r)
```
